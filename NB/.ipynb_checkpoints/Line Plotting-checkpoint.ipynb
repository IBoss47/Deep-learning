```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = np.arange(0.0, 2*math.pi, 0.1)
print(x)
```

```python
y = [math.sin(val) for val in x]
print(y)
```

### Default Line Plot
ลักษณะนี้ทาง matplotlib จะเชื่อมต่อจุดคู่ลำดับ (x0, y0), (x1, y1) ... ต่อกันไปเรื่อย ๆ ทำให้ได้เส้นที่ต่อเนื่องกันไปเป็นเนื้อเดียวกัน

```python
plt.plot(x, y)
```

### Marker Plot
ในบางครั้ง เราอยากจะแสดงเฉพาะตำแหน่งจุด ไม่ต้องการเส้น เราก็สามารถทำได้ด้วยการระบุรูปร่างและสีของจุดที่ต้องการวาด ในที่นี้เราเลือกสีแดง (red) โดยใช้ Marker เครื่องหมาย + แทนตำแหน่งจุด

```python
plt.plot(x, y, 'r+')
```

### ลองเปลี่ยนสีและลักษณะของจุด
คราวนี้เราลองเลือกสีน้ำเงิน (blue) และใช้จุดกลุมเป็นเครื่องหมายตำแหน่งจุด
มาถึงจุดนี้แล้ว เราคงจะสงสัยว่าแล้วจะรู้ได้ไงว่ามีสีอะไรและมาร์คเกอร์แบบไหนที่ใช้ได้บ้าง ถ้าใครสงสัยก็ลองดูโนัตเกี่ยวกับสีและมาร์คเกอร์ได้ที่[ลิงค์นี้](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)

และถ้าใครสงสัยว่า markdown ใน Jupyter Notebook ที่น่าสนใจมีอะไรบ้างก็ลองดูที่[ลิงค์นี้](https://medium.com/ibm-data-science-experience/markdown-for-jupyter-notebooks-cheatsheet-386c05aeebed)

```python
plt.plot(x, y, 'bo')
```

**ลองพล็อตจุดแบบห่าง ๆ กัน ซึ่งทำได้โดยการเพิ่ม step size จาก 0.1 ให้เป็น 0.2**

```python
x = np.arange(0.0, 2*math.pi, 0.2)
y = [math.sin(val) for val in x]
plt.plot(x, y, 'bo')
```

### แล้วถ้าอยากพล็อตทั้งเส้นทั้งจุดล่ะ
เท่าที่เห็น พอเรามีการระบุลักษณะของมาร์คเกอร์ลงในคำสั่ง plot ปุ๊บ เส้นก็หายปั๊บ แต่ถ้าเราอยากจะพล็อตทั้งเส้นและจุดล่ะ จะทำยังไง  ในกรณีเช่นนี้ วิธีที่นิยมที่สุดก็คือการระบุชนิดของเส้นเข้าไปด้วย พร้อมกับลักษณะของมาร์คเกอร์ ซึ่ง plot จะรู้ว่าเราต้องการทั้งเส้นและมาร์คเกอร์

```python
plt.plot(x, y, 'bo', linestyle='dashed')
```

### ไปให้สุด ตกแต่งกราฟเต็มที่
บางทีเราก็รู้สึกว่าสีของเส้นกับมาร์คเกอร์มันไม่ควรจะเหมือนกัน และเราอยากระบุความหนาของเส้น ขนาดของมาร์คเกอร์ด้วย ยิ่งไปกว่านั้น ที่จริงสีที่ระบายในตัวมาร์คเกอร์กับตัวเส้นของมาร์คเกอร์จะต่างกันก็ได้ อย่างในที่นี้ เราให้เส้นมาร์คเกอร์เป็นสีเหลือง, แต่สีไส้ในเป็นสีเขียว

```python
plt.plot(x, y, 'bo', linewidth=1, markersize=6, linestyle='dashed', 
         color='r', markerfacecolor='y', markeredgewidth=1,markeredgecolor='g')
```

### เมื่อเราถือสัญชาติไพธอน เราต้องไม่พิมพ์เยอะ เมื่อยนิ้ว
ชาวไพธอนจำนวนมากไม่ได้อยู่ในสาขาคอมพิวเตอร์ ความเร็วในการพิมพ์กับอุปนิสัยในการเขียนโค้ดแบบชื่อตัวแปรยาว ๆ จะไม่มีมากเท่าคนในสายคอม (ที่ถูกบริษัทบังคับให้เขียนโค้ดให้ชัด ๆ) ดังนั้นชื่ออะไรที่ยาว ๆ มักจะมีคำย่อให้ อย่าง markerfacecolor ก็ย่อเป็น mfc ส่วน markeredgewidth และ markeredgecolor ก็ย่อได้เป็น mew และ mec

ปล. ที่คนสอนรู้ว่ามันย่อได้ก็เพราะอ่านเอกสารของคำสั่ง plot เอานี่แหละ ไม่ใช่ว่าอยู่ดี ๆ จะรู้เองได้

```python
plt.plot(x, y, 'bo', linewidth=1, markersize=6, linestyle='dashed', 
         color='r', mfc='y', mew=1, mec='g')
```
